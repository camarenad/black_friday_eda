```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, VarianceThreshold,mutual_info_regression, f_regression
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
import warnings

warnings.filterwarnings('ignore')
np.random.seed(42)
```

```python
X_train = pd.read_pickle('../assets/X_train.pkl')
X_test = pd.read_pickle('../assets/X_test.pkl')
y_train = pd.read_pickle('../assets/y_train.pkl')
y_test = pd.read_pickle('../assets/y_test.pkl')
```

```python
pipe = Pipeline([
    ('ss',StandardScaler()),
    ('kbest',SelectKBest(f_regression,k=25)),
    ('rf_regressor',RandomForestRegressor(bootstrap=True,n_estimators=300)),
])
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.fit(X_test,y_test)
```

```python
pipe.score(X_train,y_train)
```

```python
pipe.score(X_test,y_test)
```

```python
params = {
    'rf_regressor__n_estimators':[500,700,800,1000],
#     'rf_regressor__bootstrap':[True,False],
    'rf_regressor__max_depth':[5,20,80,400,800],
    'rf_regressor__min_samples_split':[2,3,6,9,12,25]
}
```

```python
gs = GridSearchCV(pipe,
                  param_grid=params,)
```

```python
gs.fit(X_train,y_train)
```

```python
gs.fit(X_test,y_test)
```

```python
gs.score(X_train,y_train)
```

```python
gs.score(X_test,y_test)
```

```python
gs.best_params_
```

```python
X_train.columns
```

```python
mask = gs.best_estimator_.named_steps.kbest.get_support()
```

```python
X_train.columns[mask]
```

```python
feat_importances = pd.DataFrame(gs.best_estimator_.named_steps['rf_regressor'].feature_importances_, X_train.columns[mask], columns=['importance'])
```

```python
plt.figure(figsize=(20,10))
fig = sns.barplot(x='importance', 
            y=feat_importances.sort_values('importance', ascending=False).index, 
            data=feat_importances.sort_values('importance', ascending=False), 
            color='#94d1c8',
            orient='horizontal')
plt.xlabel('Importance', size='35')
plt.xticks(size=20)
plt.ylabel("Features", size='35')
plt.yticks(size=20)
plt.tight_layout()
```
